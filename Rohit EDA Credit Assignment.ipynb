### Imputing values in the Columns 

Defining a function that fills null values in different columns with the appropriate values.

```python
def fill_null_values (df,command):
    if command == "mode":
        for i in df:
            Application_DataFrame[i].fillna(Application_DataFrame[i].mode()[0],inplace = True )
    elif command == "median":
        for i in df:
            Application_DataFrame[i].fillna(Application_DataFrame[i].median(), inplace = True)
```

Numeric Columns with < 19 % null values will be imputed with mode

```python
Numeric_Cols = ['AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_DAY', 'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_QRT', 'AMT_REQ_CREDIT_BUREAU_YEAR', 'CNT_FAM_MEMBERS']
```

```python
fill_null_values (Numeric_Cols,"mode")
```

Continuous Columns with < 20 % null values will be imputed with median

```python
Cont_Cols = ['EXT_SOURCE_2','EXT_SOURCE_3','AMT_GOODS_PRICE','AMT_ANNUITY']
Cont_Cols
```
